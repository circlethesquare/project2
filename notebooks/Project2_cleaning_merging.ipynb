```python
import pandas as pd
import numpy as np
```

```python
final_web_data_p1_path = "../data/raw/df_final_web_data_pt_1.txt"
final_web_data_pt_2_path = "../data/raw/df_final_web_data_pt_2.txt"
final_experiment_clients_path = "../data/raw/df_final_experiment_clients.txt"
final_demo_path ="../data/raw/df_final_demo.txt"
```

```python
final_web_data_p1_raw = pd.read_csv(final_web_data_p1_path)
final_web_data_pt_2_raw = pd.read_csv(final_web_data_pt_2_path)
final_experiment_clients_raw = pd.read_csv(final_experiment_clients_path)
final_demo_raw = pd.read_csv(final_demo_path)
```

## Data cleaning for final_web_data

```python
#Concating two parts of the web data 
final_web_data_raw = pd.concat([final_web_data_p1_raw, final_web_data_pt_2_raw], ignore_index=True)

#checking for null values, non existing : 

final_web_data_raw.isnull().sum()

#Checking for duplicated and dropping duplicated ones 

final_web_data_raw.duplicated().sum()

#there're 10764 duplicated rows 

final_web_data_raw.drop_duplicates(keep='first', inplace = True)

#also resetting index 

final_web_data_raw.reset_index()
```

### Data cleaning for final_experiment_clients_raw

```python
final_experiment_clients_raw.isnull().sum()

#dropping null values as this is a two column data anyway 

final_experiment_clients_raw.dropna(inplace=True)

#no duplicates 
final_experiment_clients_raw.rename(columns={'Variation':'variation'})
```

## data cleaning for final_demo_raw 

```python
final_demo_raw.isnull().sum()
final_demo_raw.dropna(how='any',inplace=True)
```

```python
final_demo_raw.reset_index()
```

```python
#checking for the instersation number of clients 
intersection_client_ids = final_demo_raw['client_id'][final_demo_raw['client_id'].isin(final_web_data_raw['client_id'])]
```

```python
#merging data frames for mathing clients and keeping all other the columns  

merged_df = pd.merge(final_demo_raw, 
                     final_web_data_raw, 
                     on='client_id', 
                     how='inner'
)
```

```python
#merging data frames for mathing clients and keeping all other the columns  


merged2_df = pd.merge(merged_df, 
         final_experiment_clients_raw, 
         on='client_id', 
         how='inner'
)
```

```python
merged2_df.to_csv(path_or_buf='../data/cleaned/cleaned_client_data.csv')
```
